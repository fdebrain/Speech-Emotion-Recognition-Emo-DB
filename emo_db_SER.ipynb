# Feature Engineering & Modeling

**Exploring the potential features:** We take a look at the available features provided by Librosa API, a library commonly used for audio analysis.

```python
# Visualize ZC stats accross different emotions
df['ZC'] = [compute_zc(x) for x in  df['signal']]

fig, ax = plt.subplots(figsize=(14,5))
df.boxplot(column='ZC', by='emotion', ax=ax)
fig.suptitle('')
ax.set_title('ZC boxplots vs emotion')
ax.set_xlabel('')
ax.set_ylabel('ZC')
plt.show()
```

**Observation:** The number of zero crossings in a signal might not be discriminative enough to be a good feature.

```python
@interact(sample_id=range(0, len(df)))
def load_wav(sample_id=0):
    fig, ax = plt.subplots(6,1, figsize=(10,10))
    
    # Extract signal from db
    x = df['signal'].loc[sample_id]
    
    # Play sound
    ipd.display(ipd.Audio(data=x, rate=22050,  element_id=0))
    
    # Visualize temporal signal
    librosa.display.waveplot(x, ax=ax[0])
    
    # Feature - Extract Zero Crossing Rate
    zcr = compute_zcr(x)
    ax[1].plot(zcr)
    
    # Feature - STFT
    stft = librosa.stft(x)
    stft = librosa.amplitude_to_db(abs(stft), ref=np.max)
    librosa.display.specshow(stft, x_axis='time', y_axis='log', ax=ax[2])
    ax[2].set_title('STFT')
        
    # Feature - Mel-Frequency Cepstral Coefficients (MFCC) (describes overall shape of spectral envelope)
    mfcc = librosa.feature.mfcc(x, n_mfcc=20)
    librosa.display.specshow(mfcc, x_axis='time', ax=ax[3])
    ax[3].set_title('MFCC')

    # Feature - Chroma
    chroma = librosa.feature.chroma_stft(S=stft)
    librosa.display.specshow(chroma, x_axis='time', ax=ax[4])
    ax[4].set_title('Chroma')
    
    # Feature - Mel
    mel = librosa.feature.melspectrogram(x)
    mel = librosa.amplitude_to_db(abs(mel), ref=np.max)
    librosa.display.specshow(mel, x_axis='time', y_axis='log', ax=ax[5])
    ax[5].set_title('Mel spectrogram')

    # Extract information
    emotion, label = df['emotion'].loc[sample_id], df['label'].loc[sample_id] 
    fig.suptitle(f'Label: {label} - Emotion: {emotion}', y=1.02)
    print(f'Signal shape: {x.shape}')
    
    plt.tight_layout()
```

**Observations:** The high dimensionality of STFT and Mel features might be problematic for a first simple approach. However, MFCC and Chroma seem to have a relatively low dimensionality. In order to palliate the different length accross samples, we might need to apply a mean or padding operation.

```python
# Feature extraction (MFCC)
X_train = np.array([extract_features(x, mfcc=True) for x in df_train['signal']])
y_train = np.array(df_train['label'])
train_ids = np.array(df_train['speaker'])

X_test = np.array([extract_features(x, mfcc=True) for x in df_test['signal']])
y_test = np.array(df_test['label'])

# Scaling
mms = MinMaxScaler(feature_range=(-1, 1))
X_train = mms.fit_transform(X_train)
X_test = mms.transform(X_test)
```

```python
%matplotlib inline

# Projecting on first 2 PCs
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train)
X_pca_test = pca.transform(X_test)

fig, ax = plt.subplots(1,2, figsize=(12,6))
scatter = ax[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y_train, cmap="Spectral") #c=plt.cm.Accent(y_train.astype(int)))
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[0].set_title('Projection of train set features')
legend = ax[0].legend(scatter.legend_elements()[0], list(label2emotion.values()))
ax[0].add_artist(legend)

ax[1].scatter(X_pca_test[:, 0], X_pca_test[:, 1], c=y_test, cmap="Spectral")
ax[1].set_xlabel('PC1')
ax[1].set_ylabel('PC2')
ax[1].set_title('Projection of test set features')
```

```python
# Define base models
models = {'SVC': get_SVC(), 'RFC': get_RFC(), 'MPC': get_MPC(), 'KNNC': get_KNNC(), 'XGBC': get_XGBC()}
trained_models = []

# Optimize models based via Group K-fold CV and Bayesian optimization (~2mn)
for model_name, (model, search_space, n_iters) in models.items():
    pipeline, cv_mean, cv_std, training_time = train(X_train, y_train, train_ids, model, search_space, 'optimize', n_iters, SEED)
    trained_models.append((model_name, pipeline.best_estimator_, cv_mean))
    print(f'{model_name} fitted in {training_time:.1f}s')
    print(f'Validation F1: {cv_mean:.3f}+-{cv_std:.3f} \n')
```

```python
# Sort models by mean CV score
trained_models.sort(key=lambda l: l[-1], reverse=True)

# Extract n best models (and drop scores)
n_best = 3
weights = compute_weights(trained_models[:n_best])
base_models = [m[:2] for m in trained_models[:n_best]]

# Ensemble model using top n_best models (based on CV score)
ensemble_model, _ = get_VC(base_models, weights)

ensemble_model, y_preds, training_time = train(X_train, y_train, train_ids, ensemble_model, 
                                               search_space=None, mode='validate', seed=SEED)
print(f'Ensemble fitted in {training_time:.1f}s')
print(f'Validation F1: {score(y_train, y_preds):.3f}')
```

# Results & Visualizations

```python
# Test score for individual models
for name, model, _ in trained_models:
    print(f'{name}: {compute_score(model, X_test, y_test):.3f}')
    
# Test score for ensemble
print(f'Ensemble: {compute_score(ensemble_model, X_test, y_test):.3f}')
```

```python
%matplotlib inline

# Show confusion matrix of best base model (based on validation)
print(f'Best base model: {trained_models[0][0]}')
y_pred = trained_models[0][1].predict(X_test)
fig = compute_cm(y_test, y_pred, classes=label2emotion.values(), title='Confusion matrix - Best base model')

# Compute metrics
report = classification_report(y_test, y_pred, output_dict=True, target_names=label2emotion.values())
report_df = pd.DataFrame(report)
report_df
```

```python
%matplotlib inline

# Show confusion matrix of ensemble model
y_pred = ensemble_model.predict(X_test)
fig = compute_cm(y_test, y_pred, classes=label2emotion.values(), title='Confusion matrix - Ensemble model')

# Compute metrics
report = classification_report(y_test, y_pred, output_dict=True, target_names=label2emotion.values())
report_df = pd.DataFrame(report)
report_df
```
